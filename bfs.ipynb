<h2>What BFS is</h2>
<p>
BFS (Breadth-First Search) is a search method that finds the shortest path (fewest steps) by exploring locations level by level, starting from the beginning point.
</p>

<h2>BFS Steps (Simple)</h2>
<p>
I. Start from the initial location (start node).<br>
II. Put the start node into a queue.<br>
III. Mark the start node as visited.<br>
IV. Remove the first node from the queue.<br>
V. Check if the removed node is the goal. If yes, stop.<br>
VI. If it is not the goal, add all unvisited neighboring nodes to the queue.<br>
VII. Repeat steps IV to VI until the goal is found.
</p>

```python
from collections import deque

graph = {
    "CC": [("ML", 2), ("AO", 3), ("LA", 2), ("CF", 2)],
    "ML": [("FL", 1), ("DL", 2)],
    "FL": [("DL", 1)],
    "DL": [],
    "AO": [("HR", 1), ("FO", 2), ("ST", 2)],  # FIX: SL -> ST
    "HR": [("FO", 1)],
    "FO": [],
    "LA": [("L1", 1), ("C1", 2), ("MA", 2)],
    "L1": [("L2", 1)],
    "L2": [("L3", 1)],
    "L3": [("L4", 1)],
    "L4": [("PL", 2)],
    "C1": [("C2", 1)],
    "C2": [("C3", 1)],
    "C3": [("C4", 1)],
    "C4": [],
    "PL": [("CL", 2)],
    "CL": [("BL", 2)],
    "BL": [("SC", 3)],
    "SC": [],
    "CF": [("CC", 2), ("MA", 1), ("FE", 1), ("ST", 2)],
    "MA": [("LA", 2)],
    "FE": [("ML", 2)],
    "ST": [("AO", 2)]
}


def bfs(start, goal):
    queue = deque([[start]])
    visited = set()

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)

            for neighbor, _ in graph.get(node, []):
                new_path = path + [neighbor]
                queue.append(new_path)

    return None
# RUN
start = "CC"
goal = "PL"

result = bfs(start, goal)

print("START:", start)
print("GOAL :", goal)
print("----------------------")

if result:
    print("PATH FOUND:")
    print(" -> ".join(result))
    print("STEPS:", len(result) - 1)
else:
    print("No path found")
```
