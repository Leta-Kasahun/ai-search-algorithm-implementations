<h2>What Greedy Best-First Search is</h2>
<p>
Greedy Best-First Search is a search method that always expands the node that appears closest to the goal based on a heuristic function. It focuses only on the estimated cost to reach the goal, ignoring the actual cost traveled so far. This makes it fast but not always optimal.
</p>

<h2>Greedy Search Steps (Simple)</h2>
<p>
I. Start from the initial location (start node).<br>
II. Put the start node into a priority queue based on its heuristic value h(n).<br>
III. Remove the node with the smallest heuristic value (closest to goal).<br>
IV. Check if it is the goal. If yes, stop and return the path.<br>
V. If it is not the goal, explore all its neighbors.<br>
VI. For each neighbor, calculate its heuristic value h(neighbor) = estimated cost to goal.<br>
VII. Add all unvisited neighbors to the priority queue with their h values.<br>
VIII. Repeat steps III to VII until the goal is found.
</p>

```python
from graph import graph
import heapq

def heuristic(node, goal):
    heuristic_values = {
        "Physics Lab": 0,
        "Chemistry Lab": 2,
        "Biology Lab": 4,
        "Sports Center": 7,
        "LH04": 2,
        "LH03": 3,
        "LH02": 4,
        "LH01": 5,
        "Learning Area": 6,
        "CL04": 5,
        "CL03": 4,
        "CL02": 3,
        "CL01": 2,
        "Campus Center": 8,
        "Main Library": 7,
        "Admin Office": 6,
        "Café": 7,
        "Female Library": 8,
        "Digital Library": 9,
        "HR Office": 7,
        "Freshman Office": 6,
        "Staff Lounge": 5,
        "Male Lounge": 5,
        "Female Lounge": 6
    }
    return heuristic_values.get(node, 10)

def greedy_search(start, goal):
    priority_queue = [(heuristic(start, goal), start, [start])]
    visited = set()
    
    while priority_queue:
        h, node, path = heapq.heappop(priority_queue)
        
        if node in visited:
            continue
        
        visited.add(node)
        
        if node == goal:
            return path
        
        for neighbor, cost in graph[node]:
            if neighbor not in visited:
                h_value = heuristic(neighbor, goal)
                heapq.heappush(priority_queue, (h_value, neighbor, path + [neighbor]))
    
    return None

start = "Campus Center"
goal = "Physics Lab"

result = greedy_search(start, goal)

print("START:", start)
print("GOAL :", goal)
print("----------------------")

if result:
    print("PATH FOUND:")
    print(" -> ".join(result))
    print("STEPS:", len(result) - 1)
else:
    print("No path found")
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
import heapq
import time

from graph import graph

pos = {
    "Campus Center": (400, 30),
    "Main Library": (180, 120),
    "Admin Office": (350, 120),
    "Learning Area": (550, 120),
    "Café": (750, 120),
    "Female Library": (130, 220),
    "Digital Library": (260, 220),
    "HR Office": (370, 220),
    "Freshman Office": (470, 220),
    "Staff Lounge": (560, 220),
    "LH01": (640, 220),
    "CL01": (720, 220),
    "Male Lounge": (810, 220),
    "LH02": (640, 320),
    "LH03": (640, 400),
    "LH04": (640, 480),
    "Physics Lab": (640, 560),
    "Chemistry Lab": (640, 620),
    "Biology Lab": (640, 680),
    "Sports Center": (640, 740),
    "CL02": (720, 320),
    "CL03": (720, 400),
    "CL04": (720, 480),
    "Female Lounge": (900, 220)
}

def heuristic(node, goal):
    heuristic_values = {
        "Physics Lab": 0,
        "Chemistry Lab": 2,
        "Biology Lab": 4,
        "Sports Center": 7,
        "LH04": 2,
        "LH03": 3,
        "LH02": 4,
        "LH01": 5,
        "Learning Area": 6,
        "CL04": 5,
        "CL03": 4,
        "CL02": 3,
        "CL01": 2,
        "Campus Center": 8,
        "Main Library": 7,
        "Admin Office": 6,
        "Café": 7,
        "Female Library": 8,
        "Digital Library": 9,
        "HR Office": 7,
        "Freshman Office": 6,
        "Staff Lounge": 5,
        "Male Lounge": 5,
        "Female Lounge": 6
    }
    return heuristic_values.get(node, 10)

def calculate_path_cost(path):
    total = 0
    for i in range(len(path) - 1):
        for neighbor, cost in graph[path[i]]:
            if neighbor == path[i + 1]:
                total += cost
                break
    return total

def greedy_search(start, goal, callback=None):
    priority_queue = [(heuristic(start, goal), start, [start])]
    visited = set()
    
    while priority_queue:
        h, node, path = heapq.heappop(priority_queue)
        
        if node in visited:
            continue
        
        visited.add(node)
        
        if callback:
            callback(node, "exploring", h)
            time.sleep(0.3)
        
        if node == goal:
            if callback:
                for i, p in enumerate(path):
                    callback(p, "final", heuristic(p, goal))
                    time.sleep(0.15)
            return path, calculate_path_cost(path)
        
        for neighbor, cost in graph[node]:
            if neighbor not in visited:
                h_value = heuristic(neighbor, goal)
                heapq.heappush(priority_queue, (h_value, neighbor, path + [neighbor]))
                if callback:
                    callback(neighbor, "queued", h_value)
                    time.sleep(0.1)
    
    return None, None

class GreedyVisualization:
    def __init__(self, root):
        self.root = root
        self.root.title("GREEDY BEST-FIRST SEARCH - Campus Navigation")
        self.root.geometry("1200x800")
        self.root.configure(bg="#0a0a1a")
        
        main_frame = tk.Frame(root, bg="#0a0a1a")
        main_frame.pack(fill=tk.BOTH, expand=True)
        
        self.canvas = tk.Canvas(main_frame, width=900, height=780, bg="#0f0f1a", highlightthickness=0)
        self.canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=True, padx=10, pady=10)
        
        control_frame = tk.Frame(main_frame, bg="#1a1a2e", width=280)
        control_frame.pack(side=tk.RIGHT, fill=tk.Y, padx=5, pady=10)
        control_frame.pack_propagate(False)
        
        title = tk.Label(control_frame, text="GREEDY BEST-FIRST SEARCH", font=("Arial", 14, "bold"), bg="#1a1a2e", fg="#e94560")
        title.pack(pady=15)
        
        desc = tk.Label(control_frame, text="Always expands node closest to goal", font=("Arial", 9), bg="#1a1a2e", fg="#aaaaaa")
        desc.pack(pady=5)
        
        desc2 = tk.Label(control_frame, text="Uses only heuristic h(n)", font=("Arial", 8, "bold"), bg="#1a1a2e", fg="#ffd700")
        desc2.pack(pady=2)
        
        start_frame = tk.Frame(control_frame, bg="#1a1a2e")
        start_frame.pack(pady=10, fill=tk.X, padx=20)
        
        tk.Label(start_frame, text="START:", font=("Arial", 11, "bold"), bg="#1a1a2e", fg="#00ff88").pack(anchor=tk.W)
        self.start_var = tk.StringVar(value="Campus Center")
        self.start_combo = ttk.Combobox(start_frame, textvariable=self.start_var, values=list(pos.keys()), font=("Arial", 10))
        self.start_combo.pack(fill=tk.X, pady=5)
        
        goal_frame = tk.Frame(control_frame, bg="#1a1a2e")
        goal_frame.pack(pady=10, fill=tk.X, padx=20)
        
        tk.Label(goal_frame, text="GOAL:", font=("Arial", 11, "bold"), bg="#1a1a2e", fg="#e94560").pack(anchor=tk.W)
        self.goal_var = tk.StringVar(value="Physics Lab")
        self.goal_combo = ttk.Combobox(goal_frame, textvariable=self.goal_var, values=list(pos.keys()), font=("Arial", 10))
        self.goal_combo.pack(fill=tk.X, pady=5)
        
        self.run_button = tk.Button(control_frame, text="RUN GREEDY SEARCH", font=("Arial", 11, "bold"), bg="#e94560", fg="white", command=self.run_search)
        self.run_button.pack(pady=15, padx=20, fill=tk.X)
        
        self.reset_button = tk.Button(control_frame, text="RESET", font=("Arial", 10), bg="#5c5c8a", fg="white", command=self.reset_canvas)
        self.reset_button.pack(pady=5, padx=20, fill=tk.X)
        
        info_frame = tk.Frame(control_frame, bg="#0f0f1a", relief=tk.RAISED, bd=2)
        info_frame.pack(pady=15, padx=20, fill=tk.X)
        
        self.cost_label = tk.Label(info_frame, text="TOTAL COST: --", font=("Arial", 11, "bold"), bg="#0f0f1a", fg="#ffd700")
        self.cost_label.pack(pady=5)
        
        self.steps_label = tk.Label(info_frame, text="STEPS: --", font=("Arial", 11), bg="#0f0f1a", fg="#ffffff")
        self.steps_label.pack(pady=5)
        
        self.h_value_label = tk.Label(info_frame, text="h(n) = --", font=("Arial", 10), bg="#0f0f1a", fg="#88aaff")
        self.h_value_label.pack(pady=3)
        
        path_frame = tk.Frame(control_frame, bg="#1a1a2e")
        path_frame.pack(pady=10, padx=20, fill=tk.BOTH, expand=True)
        
        tk.Label(path_frame, text="PATH FOUND:", font=("Arial", 10, "bold"), bg="#1a1a2e", fg="#00ff88").pack(anchor=tk.W)
        
        self.path_text = tk.Text(path_frame, height=18, width=30, bg="#0f0f1a", fg="#e0e0ff", font=("Arial", 9), wrap=tk.WORD)
        self.path_text.pack(fill=tk.BOTH, expand=True, pady=5)
        
        self.draw_graph()
    
    def draw_graph(self):
        self.canvas.delete("all")
        for node in graph:
            if node not in pos:
                continue
            x, y = pos[node]
            text_width = len(node) * 7
            rect_x1 = x - text_width//2 - 10
            rect_x2 = x + text_width//2 + 10
            rect_y1 = y - 18
            rect_y2 = y + 18
            
            self.canvas.create_rectangle(rect_x1, rect_y1, rect_x2, rect_y2, fill="#2d2d44", outline="#5c5c8a", width=2, tags=f"node_{node}")
            self.canvas.create_text(x, y - 5, text=node, font=("Arial", 7, "bold"), fill="#e0e0ff", tags=f"text_{node}")
            
            h_val = heuristic(node, "Physics Lab")
            self.canvas.create_text(x, y + 8, text=f"h={h_val}", font=("Arial", 6), fill="#ffd700", tags=f"heuristic_{node}")
            
            for neighbor, weight in graph[node]:
                if neighbor in pos:
                    nx, ny = pos[neighbor]
                    self.canvas.create_line(x, y, nx, ny, fill="#3d3d5c", width=2)
                    mid_x, mid_y = (x + nx)//2, (y + ny)//2
                    self.canvas.create_text(mid_x, mid_y - 5, text=str(weight), font=("Arial", 8), fill="#ffd700", tags=f"weight_{node}_{neighbor}")
    
    def reset_canvas(self):
        self.draw_graph()
        self.cost_label.config(text="TOTAL COST: --")
        self.steps_label.config(text="STEPS: --")
        self.h_value_label.config(text="h(n) = --")
        self.path_text.delete(1.0, tk.END)
        self.run_button.config(state=tk.NORMAL, text="RUN GREEDY SEARCH")
    
    def update_node_with_h(self, node, color, h_cost):
        if node not in pos:
            return
        x, y = pos[node]
        text_width = len(node) * 7
        rect_x1 = x - text_width//2 - 10
        rect_x2 = x + text_width//2 + 10
        rect_y1 = y - 18
        rect_y2 = y + 18
        
        self.canvas.delete(f"node_{node}")
        self.canvas.create_rectangle(rect_x1, rect_y1, rect_x2, rect_y2, fill=color, outline="#7a7aaa", width=2, tags=f"node_{node}")
        self.canvas.create_text(x, y - 5, text=node, font=("Arial", 7, "bold"), fill="#ffffff", tags=f"text_{node}")
        self.canvas.create_text(x, y + 8, text=f"h={h_cost}", font=("Arial", 6), fill="#ffd700", tags=f"heuristic_{node}")
        
        self.root.update()
    
    def animate_smooth_path(self, path, total_cost):
        steps = len(path)
        time_per_node = 2.5 / steps
        
        for idx, node in enumerate(path):
            current_cost = calculate_path_cost(path[:idx+1])
            h_val = heuristic(node, self.goal_var.get())
            
            x, y = pos[node]
            text_width = len(node) * 7
            rect_x1 = x - text_width//2 - 10
            rect_x2 = x + text_width//2 + 10
            rect_y1 = y - 18
            rect_y2 = y + 18
            
            for i in range(15):
                alpha = i / 15
                if i < 8:
                    r, g, b = 50, 50, 100
                else:
                    r, g, b = 0, 255, 68
                
                color = f'#{int(r + (0 - r) * alpha):02x}{int(g + (255 - g) * alpha):02x}{int(b + (68 - b) * alpha):02x}'
                
                self.canvas.delete(f"node_{node}")
                self.canvas.create_rectangle(rect_x1, rect_y1, rect_x2, rect_y2, fill=color, outline="#7a7aaa", width=2, tags=f"node_{node}")
                self.canvas.create_text(x, y - 5, text=node, font=("Arial", 7, "bold"), fill="#ffffff", tags=f"text_{node}")
                self.canvas.create_text(x, y + 8, text=f"h={h_val}", font=("Arial", 6), fill="#ffd700", tags=f"heuristic_{node}")
                self.root.update()
                time.sleep(time_per_node / 15)
            
            self.cost_label.config(text=f"TOTAL COST: {current_cost}")
            self.h_value_label.config(text=f"h({node}) = {h_val}")
            self.root.update()
        
        for node in path:
            x, y = pos[node]
            text_width = len(node) * 7
            rect_x1 = x - text_width//2 - 10
            rect_x2 = x + text_width//2 + 10
            rect_y1 = y - 18
            rect_y2 = y + 18
            self.canvas.delete(f"node_{node}")
            self.canvas.create_rectangle(rect_x1, rect_y1, rect_x2, rect_y2, fill="#00ff44", outline="#7a7aaa", width=2, tags=f"node_{node}")
            self.canvas.create_text(x, y - 5, text=node, font=("Arial", 7, "bold"), fill="#000000", tags=f"text_{node}")
            self.root.update()
            time.sleep(0.08)
        
        self.cost_label.config(text=f"TOTAL COST: {total_cost}")
    
    def display_path(self, path, cost):
        self.path_text.delete(1.0, tk.END)
        for i, node in enumerate(path):
            h_val = heuristic(node, self.goal_var.get())
            
            if i == 0:
                self.path_text.insert(tk.END, f"START: {node}\n")
                self.path_text.insert(tk.END, f"       h={h_val}\n\n")
            elif i == len(path) - 1:
                self.path_text.insert(tk.END, f"GOAL: {node}\n")
                self.path_text.insert(tk.END, f"       h={h_val}\n")
                self.path_text.insert(tk.END, f"\nTOTAL COST: {cost}\n")
            else:
                self.path_text.insert(tk.END, f"{node}\n")
                self.path_text.insert(tk.END, f"       h={h_val}\n\n")
        
        self.cost_label.config(text=f"TOTAL COST: {cost}")
        self.steps_label.config(text=f"STEPS: {len(path) - 1}")
    
    def search_callback(self, node, status, h_cost):
        if status == "exploring":
            self.update_node_with_h(node, "#ffaa44", h_cost)
            self.h_value_label.config(text=f"h({node}) = {h_cost}")
            time.sleep(0.25)
        elif status == "queued":
            self.update_node_with_h(node, "#88aaff", h_cost)
            time.sleep(0.1)
        elif status == "final":
            self.update_node_with_h(node, "#00ff44", h_cost)
            self.h_value_label.config(text=f"h({node}) = {h_cost}")
            time.sleep(0.12)
    
    def run_search(self):
        self.reset_canvas()
        self.run_button.config(state=tk.DISABLED, text="SEARCHING...")
        
        start = self.start_var.get()
        goal = self.goal_var.get()
        
        if start not in graph or goal not in graph:
            messagebox.showerror("Error", "Invalid location!")
            self.run_button.config(state=tk.NORMAL, text="RUN GREEDY SEARCH")
            return
        
        self.root.update()
        
        path, cost = greedy_search(start, goal, self.search_callback)
        
        if path:
            self.display_path(path, cost)
            self.animate_smooth_path(path, cost)
            self.run_button.config(state=tk.NORMAL, text="RUN GREEDY SEARCH")
        else:
            self.path_text.delete(1.0, tk.END)
            self.path_text.insert(tk.END, f"NO PATH FOUND\n")
            self.path_text.insert(tk.END, f"From '{start}' to '{goal}'")
            self.run_button.config(state=tk.NORMAL, text="RUN GREEDY SEARCH")

if __name__ == "__main__":
    root = tk.Tk()
    app = GreedyVisualization(root)
    root.mainloop()
```
