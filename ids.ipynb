<h2>What Iterative Deepening Search (IDS) is</h2>
<p>
Iterative Deepening Search combines the benefits of BFS and DFS. It repeatedly runs Depth-Limited Search with increasing depth limits until the goal is found. IDS uses the memory efficiency of DFS while guaranteeing the shortest path like BFS. It is ideal for large search spaces where the depth of the goal is unknown.
</p>

<h2>Iterative Deepening Search Steps (Simple)</h2>
<p>
I. Start with depth limit = 0.<br>
II. Perform Depth-Limited Search (DLS) from the start node with current depth limit.<br>
III. If the goal is found, return the path and stop.<br>
IV. If the goal is not found, increase depth limit by 1.<br>
V. Repeat steps II to IV until the goal is found or all nodes are exhausted.<br>
VI. Keep track of explored nodes across iterations to avoid infinite loops.<br>
VII. IDS is complete and optimal for unweighted graphs.
</p>
