<h2>What DFS is</h2>
<p>
DFS (Depth-First Search) is a search method that explores one path as deeply as possible before going back and trying another path. It goes deep first instead of exploring level by level.
</p>

<h2>DFS Steps (Simple)</h2>
<p>
I. Start from the initial location (start node).<br>
II. Put the start node into a stack (or use recursion).<br>
III. Mark the start node as visited.<br>
IV. Take the last added node (top of stack).<br>
V. Check if it is the goal. If yes, stop.<br>
VI. If it is not the goal, go to one unvisited neighbor and continue deeper.<br>
VII. If no unvisited neighbors are available, go back (backtrack).<br>
VIII. Repeat steps IV to VII until the goal is found.
</p>

```python
from graph import graph

# -----------------------------
# DFS FUNCTION (RECURSIVE)
# -----------------------------
def dfs(node, goal, visited=None, path=None):
    if visited is None:
        visited = []
    if path is None:
        path = []

    # add current node to path and visited
    visited.append(node)
    path = path + [node]

    # GOAL CHECK
    if node == goal:
        return path

    # explore neighbors deeply
    for neighbor, _ in graph[node]:
        if neighbor not in visited:
            result = dfs(neighbor, goal, visited, path)
            if result is not None:
                return result

    return None


# -----------------------------
# RUN DFS
# -----------------------------
start = "Campus Center"
goal = "Physics Lab"

result = dfs(start, goal)

# -----------------------------
# OUTPUT
# -----------------------------
print("START:", start)
print("GOAL :", goal)
print("----------------------")

if result:
    print("PATH FOUND:")
    print(" -> ".join(result))
    print("STEPS:", len(result) - 1)
else:
    print("No path found")
```

```python
import tkinter as tk
from collections import deque
import time

from graph import graph

pos = {
    "Campus Center": (400, 30),
    "Main Library": (180, 120),
    "Admin Office": (350, 120),
    "Learning Area": (550, 120),
    "Café": (750, 120),
    "Female Library": (130, 220),
    "Digital Library": (260, 220),
    "HR Office": (370, 220),
    "Freshman Office": (470, 220),
    "Staff Lounge": (560, 220),
    "LH01": (640, 220),
    "CL01": (720, 220),
    "Male Lounge": (810, 220),
    "LH02": (640, 320),
    "LH03": (640, 400),
    "LH04": (640, 480),
    "Physics Lab": (640, 560),
    "Chemistry Lab": (640, 620),
    "Biology Lab": (640, 680),
    "Sports Center": (640, 740),
    "CL02": (720, 320),
    "CL03": (720, 400),
    "CL04": (720, 480),
    "Female Lounge": (900, 220)
}

root = tk.Tk()
root.title("Campus BFS & DFS Visualization")
root.geometry("1100x850")

main_frame = tk.Frame(root)
main_frame.pack(fill=tk.BOTH, expand=True)

canvas = tk.Canvas(main_frame, width=1000, height=780, bg="#1a1a2e")
canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

info_frame = tk.Frame(main_frame, bg="#16213e", width=280)
info_frame.pack(side=tk.RIGHT, fill=tk.Y)
info_frame.pack_propagate(False)

title_label = tk.Label(info_frame, text="PATH FINDER", 
                       font=("Arial", 14, "bold"), bg="#16213e", fg="#e94560")
title_label.pack(pady=10)

algo_label = tk.Label(info_frame, text="Algorithm: BFS", 
                      font=("Arial", 12, "bold"), bg="#16213e", fg="#ff6b6b")
algo_label.pack(pady=5)

cost_label = tk.Label(info_frame, text="Path Cost: 0", 
                      font=("Arial", 12), bg="#16213e", fg="#ffffff")
cost_label.pack(pady=10)

steps_label = tk.Label(info_frame, text="Steps: 0", 
                       font=("Arial", 12), bg="#16213e", fg="#ffffff")
steps_label.pack(pady=5)

path_title = tk.Label(info_frame, text="EXPLORED PATH:", 
                      font=("Arial", 11, "bold"), bg="#16213e", fg="#00ff88")
path_title.pack(pady=5)

path_text = tk.Text(info_frame, height=25, width=32, bg="#1a1a2e", 
                    fg="#ffffff", font=("Courier", 9), wrap=tk.WORD)
path_text.pack(padx=10, pady=10, fill=tk.BOTH, expand=True)

button_frame = tk.Frame(info_frame, bg="#16213e")
button_frame.pack(pady=10, fill=tk.X)

bfs_button = tk.Button(button_frame, text="RUN BFS", 
                       font=("Arial", 11, "bold"), bg="#e94560", fg="white",
                       command=lambda: run_bfs())
bfs_button.pack(pady=5, padx=10, fill=tk.X)

dfs_button = tk.Button(button_frame, text="RUN DFS", 
                       font=("Arial", 11, "bold"), bg="#00ff88", fg="#1a1a2e",
                       command=lambda: run_dfs())
dfs_button.pack(pady=5, padx=10, fill=tk.X)

reset_button = tk.Button(button_frame, text="RESET", 
                         font=("Arial", 11, "bold"), bg="#5c5c8a", fg="white",
                         command=lambda: reset_canvas())
reset_button.pack(pady=5, padx=10, fill=tk.X)

def get_text_width(text):
    return len(text) * 7

def draw_graph():
    canvas.delete("all")
    for node in graph:
        x, y = pos[node]
        text_width = get_text_width(node)
        rect_x1 = x - text_width//2 - 10
        rect_x2 = x + text_width//2 + 10
        rect_y1 = y - 12
        rect_y2 = y + 12
        
        canvas.create_rectangle(rect_x1, rect_y1, rect_x2, rect_y2, 
                                fill="#2d2d44", outline="#5c5c8a", width=2,
                                tags=node)
        canvas.create_text(x, y, text=node, font=("Arial", 8, "bold"), 
                          fill="#e0e0ff", tags=f"text_{node}")
        
        for neighbor, weight in graph[node]:
            if neighbor in pos:
                nx, ny = pos[neighbor]
                canvas.create_line(x, y, nx, ny, fill="#3d3d5c", width=1)

def reset_canvas():
    draw_graph()
    cost_label.config(text="Path Cost: 0")
    steps_label.config(text="Steps: 0")
    path_text.delete(1.0, tk.END)

def animate_color_change(x, y, target_color, node_text, steps=15):
    text_width = get_text_width(node_text)
    rect_x1 = x - text_width//2 - 10
    rect_x2 = x + text_width//2 + 10
    rect_y1 = y - 12
    rect_y2 = y + 12
    
    r1, g1, b1 = 45, 45, 68
    r2, g2, b2 = int(target_color[1:3], 16), int(target_color[3:5], 16), int(target_color[5:7], 16)
    
    for i in range(steps + 1):
        r = int(r1 + (r2 - r1) * i / steps)
        g = int(g1 + (g2 - g1) * i / steps)
        b = int(b1 + (b2 - b1) * i / steps)
        color = f'#{r:02x}{g:02x}{b:02x}'
        
        canvas.delete(node_text)
        canvas.delete(f"text_{node_text}")
        canvas.create_rectangle(rect_x1, rect_y1, rect_x2, rect_y2, 
                                fill=color, outline="#7a7aaa", width=2,
                                tags=node_text)
        canvas.create_text(x, y, text=node_text, 
                          font=("Arial", 8, "bold"), fill="#ffffff",
                          tags=f"text_{node_text}")
        root.update()
        time.sleep(0.03)

def update_info_panel(path, total_cost, algo_name):
    algo_label.config(text=f"Algorithm: {algo_name}")
    cost_label.config(text=f"Path Cost: {total_cost}")
    steps_label.config(text=f"Steps: {len(path) - 1}")
    path_text.delete(1.0, tk.END)
    
    for i, node in enumerate(path):
        if i == 0:
            path_text.insert(tk.END, f"START: {node}\n", "start")
        elif i == len(path) - 1:
            path_text.insert(tk.END, f"GOAL: {node}\n", "goal")
        else:
            path_text.insert(tk.END, f"  {i}. {node}\n", "normal")
    
    path_text.tag_config("start", foreground="#ff6b6b", font=("Courier", 9, "bold"))
    path_text.tag_config("goal", foreground="#00ff88", font=("Courier", 9, "bold"))
    path_text.tag_config("normal", foreground="#e0e0ff")

def run_bfs():
    reset_canvas()
    start = "Campus Center"
    goal = "Physics Lab"
    
    queue = deque([[start]])
    visited = set()
    explored_color = "#e94560"
    path_color = "#00ff88"
    queue_color = "#ff6b6b"
    
    while queue:
        path = queue.popleft()
        node = path[-1]
        x, y = pos[node]
        
        animate_color_change(x, y, queue_color, node)
        
        if node == goal:
            total_cost = 0
            for i, p in enumerate(path):
                px, py = pos[p]
                animate_color_change(px, py, path_color, p)
                
                if i > 0:
                    for neighbor, weight in graph[path[i-1]]:
                        if neighbor == p:
                            total_cost += weight
                            break
                
                update_info_panel(path[:i+1], total_cost, "BFS")
            
            return path
        
        if node not in visited:
            visited.add(node)
            animate_color_change(x, y, explored_color, node)
            
            temp_cost = 0
            for i in range(1, len(path)):
                for neighbor, weight in graph[path[i-1]]:
                    if neighbor == path[i]:
                        temp_cost += weight
                        break
            update_info_panel(path, temp_cost, "BFS")
            
            for neighbor, _ in graph[node]:
                if neighbor not in visited:
                    queue.append(path + [neighbor])

def dfs_recursive(node, goal, visited=None, path=None):
    if visited is None:
        visited = []
    if path is None:
        path = []
    
    visited.append(node)
    path = path + [node]
    
    x, y = pos[node]
    animate_color_change(x, y, "#ff6b6b", node)
    
    temp_cost = 0
    for i in range(1, len(path)):
        for neighbor, weight in graph[path[i-1]]:
            if neighbor == path[i]:
                temp_cost += weight
                break
    update_info_panel(path, temp_cost, "DFS")
    
    if node == goal:
        total_cost = 0
        for i, p in enumerate(path):
            px, py = pos[p]
            animate_color_change(px, py, "#00ff88", p)
            
            if i > 0:
                for neighbor, weight in graph[path[i-1]]:
                    if neighbor == p:
                        total_cost += weight
                        break
            
            update_info_panel(path[:i+1], total_cost, "DFS")
        
        return path
    
    animate_color_change(x, y, "#e94560", node)
    
    for neighbor, _ in graph[node]:
        if neighbor not in visited:
            result = dfs_recursive(neighbor, goal, visited, path)
            if result is not None:
                return result
    
    return None

def run_dfs():
    reset_canvas()
    start = "Campus Center"
    goal = "Physics Lab"
    
    result = dfs_recursive(start, goal)
    
    if result:
        print("DFS Path:", " -> ".join(result))
    else:
        print("No path found")

draw_graph()
root.mainloop()
```
