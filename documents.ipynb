<h2>Campus Navigation System</h2> <p> This system solves the real-world problem of finding efficient routes between different locations inside a campus. In a large campus environment, users often struggle to move between buildings such as libraries, laboratories, lecture halls, departments, and administrative offices. The system models this environment as a search problem and applies classical AI search algorithms to determine the best possible path between two locations. </p>
<h2>What Problem It Solves</h2> <p> The system solves the problem of inefficient navigation inside a campus where users may waste time or get confused when moving between buildings. It provides an intelligent way to find optimal routes based on distance, cost, or number of steps depending on the search strategy used. </p>
<h2>How It Solves the Problem</h2> <p> The campus is represented as a graph where each location is a node and each path between locations is an edge with an associated cost. Search algorithms explore this graph differently to find a route from a starting location to a destination. Breadth-First Search explores level by level, Depth-First Search explores deeply, Uniform Cost Search selects the least-cost path, Greedy Search uses heuristic direction toward the goal, A* combines cost and heuristic for optimal performance, and Bidirectional Search reduces search time by exploring from both ends. </p>
<h2>Problem Formulation</h2> <h3>1. State Space</h3> <p> The state space consists of all possible locations in the campus such as lecture halls, laboratories, libraries, departments, and offices. Each location represents a state that the search algorithm can visit. </p> <h3>2. Initial State</h3> <p> The initial state is the starting location where the user begins navigation. </p> <h3>3. Goal State</h3> <p> The goal state is the destination location the user wants to reach. </p> <h3>4. Actions (Operators)</h3> <p> Actions represent movement from one location to another through connected paths in the campus graph. </p> <h3>5. Path Cost</h3> <p> Path cost represents the numerical value assigned to each movement between locations. It may represent distance, time, or difficulty of travel. </p>
<h2>Search Problem Components</h2> <h3>1. State</h3> <p> Each node in the graph represents a campus location such as lecture halls, labs, or offices. </p> <h3>2. Initial State</h3> <p> The starting point of the search process. </p> <h3>3. Goal State</h3> <p> The destination node that the system must reach. </p> <h3>4. Actions</h3> <p> Possible transitions between connected nodes in the graph. </p> <h3>5. Path Cost</h3> <p> The cost assigned to each movement used to calculate optimal paths. </p>
<h2>Summary of Search Logic</h2> <p> The system uses different search algorithms to explore the campus graph and find efficient routes. BFS ensures the shortest path in terms of steps, DFS explores deeply, UCS finds the lowest-cost route, Greedy Search moves toward the goal quickly using heuristics, A* provides the best balance between cost and heuristic, and Bidirectional Search improves efficiency by searching from both start and goal simultaneously. </p>
